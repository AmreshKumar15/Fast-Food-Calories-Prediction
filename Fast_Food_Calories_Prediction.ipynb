# SMOTE - FOR BALANCING THE OUTCOMES

```python
# Initial class distribution
print("Class distribution before SMOTE:")
print(y.value_counts())

# Apply SMOTE to oversample the minority class
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_train, y = smote.fit_resample(X_train, y)

# Class distribution after SMOTE
print("\nClass distribution after SMOTE:")
print(pd.Series(y).value_counts())
```

```python
df_train = pd.concat([X_train, y], axis=1)
sns.countplot(x='Healthy', data=df_train)
plt.title('Healthy Count Plot')
plt.show()
```

# MODEL SELECTION USING CROSS VALIDATION

```python
# Define models
models = [
    RandomForestClassifier(n_estimators=300, random_state = 42),
    LogisticRegression(max_iter=1000, random_state = 42),
    SVC(random_state = 42),
    DecisionTreeClassifier(random_state = 42),
    AdaBoostClassifier(n_estimators=400, random_state = 42),
    GradientBoostingClassifier(n_estimators=400, random_state = 42),
    KNeighborsClassifier(n_neighbors=5),
    GaussianNB()
]
print("MEAN ACCURACIES:")
# Loop through each model
for model in models:
    # Perform cross-validation
    cv_scores = cross_val_score(model, X_train, y, cv=5, scoring='accuracy')
    # Print average accuracy for each model
    print(f'{model.__class__.__name__}: {np.mean(cv_scores):.4f} (+/- {np.std(cv_scores):.4f})')
```
